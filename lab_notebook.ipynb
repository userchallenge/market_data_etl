```python
omx30 = pd.read_csv("omxs30_original.csv")[
    ["Date", "Open", "High", "Low", "Close", "Volume"]
]

# Replace all commas with dots in the entire dataframe
omx30 = omx30.applymap(lambda x: x.replace(",", ".") if isinstance(x, str) else x)

omx30 = omx30.rename(
    columns={
        "Date": "date",
        "Open": "open",
        "High": "high",
        "Low": "low",
        "Close": "close",
        "Volume": "volume",
    }
)

# remove time-component from date
omx30["date"] = omx30["date"].str.split(" ").str[0]


print(omx30.info())
omx30.to_csv(
    "omxs30_cleaned.csv", index=False
)  # Save the cleaned data to a new CSV file
omx30
```
